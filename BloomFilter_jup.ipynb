```python
#BloomFilter_jup.ipynb
import matplotlib.pyplot as plt
from BloomFilter import BloomFilter
```

```python
# Initialize a Bloom Filter
bf = BloomFilter(1000, 0.01)
```

```python
# Add items to the Bloom Filter
items = ["apple", "banana", "grape", "cherry"]
for item in items:
    bf.add(item)
```

```python
# Check for presence of items
check_items = ["apple", "banana", "mango", "grape"]
presence = {item: bf.contains(item) for item in check_items}
```

```python
# Display results
print("Bloom Filter Check Results:")
for item, is_present in presence.items():
    print(f"{item}: {'Present' if is_present else 'Not Present'}")
```

```python

# Plotting example
plt.bar(presence.keys(), presence.values())
plt.xlabel('Items')
plt.ylabel('Presence in Bloom Filter')
plt.title('Bloom Filter Presence Check')
plt.show()
```
